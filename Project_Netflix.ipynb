```python
# Save in movies by release year
# Group by the year of release
# Get the row count for each group
# Sort the values in descending order
# Reset index and create new colum labeled Movie Count

movies_by_release_year = movies.groupby('release_year').size()\
                               .sort_values(ascending = False)\
                               .reset_index(name="movie_count")
movies_by_release_year
```

```python
# Movies released on and prior to 1979
movies_before_1980 = movies_by_release_year[movies_by_release_year['release_year']<=1979]

# Movies released betweem 1980 and 1999
movies_between_1980_2000 = movies_by_release_year[(movies_by_release_year['release_year'] >= 1980) & (movies_by_release_year['release_year'] < 2000)]

# Movies released on and after 2000
movies_after_2000 = movies_by_release_year[movies_by_release_year['release_year']>=2000]
```

```python
#Sort all according release year

before_1979 = movies_before_1980.sort_values('release_year',ascending = False)

between_1980_2000 = movies_between_1980_2000.sort_values('release_year',ascending=False)

after_2000 = movies_after_2000.sort_values('release_year',ascending=False)
```

```python
plt.plot(before_1979.release_year,\
        before_1979.movie_count,\
        label = "Until 1979",\
        linestyle = "--",\
        marker="o",\
        color = "blue")

plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Movies by release year")
plt.legend()
plt.show()
```

```python
plt.plot(between_1980_2000.release_year,\
         between_1980_2000.movie_count,\
         label = "From 1980 to 2000",\
         linestyle = "--",\
         marker = "o",\
         color = "green")

plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Movies by release year")
plt.legend()
plt.show()
```

```python
plt.plot(after_2000.release_year,\
         after_2000.movie_count,\
         label = "Movies after 2000",\
         marker = "o",\
         linestyle = "--",\
         color = "r")

plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Movies by release year")
plt.legend()
plt.show()











```

```python
# Remove all nan values from `cast` column

noNanValuesDf = importedData.dropna(subset=['cast'])
noNanValuesDf
```

```python
movieCast = noNanValuesDf.loc[noNanValuesDf['type']=="Movie"]

tvshowCast = noNanValuesDf.loc[noNanValuesDf['type']=="TV Show"]
```

```python
#Saave to splitMovieCast
# Set the Index to an existing column (type)

splitMovieCast = movieCast.set_index('type')\
                 .cast.str.split(",",expand=True)\
                .stack()\
                 .reset_index('type')\
                 
splitMovieCast=splitMovieCast.rename(columns = {0 : 'Actor'})
splitMovieCast
```

```python
#Swaps Column
splitMovieCast = splitMovieCast.sort_values('Actor')[['Actor','type']]
splitMovieCast
```

```python
splitMovieCast = splitMovieCast.reset_index(drop=True)
splitMovieCast
```

```python
moviePerActor = splitMovieCast.groupby('Actor').size()\
                              .sort_values(ascending = False)\
                              .reset_index(name = "Sum of Movies")
moviePerActor
```

```python
#TV Show Actors
splitTVShowCast = tvshowCast.set_index('type')\
                            .cast.str.split(",",expand= True)\
                            .stack()\
                            .reset_index('type')\
                            .rename(columns = {0 : "Actor"})\
                            .sort_values('Actor')[['Actor','type']]\
                            .reset_index(drop = True)
splitTVShowCast
```

```python
#TV Show per Actor
tvshowPerActor = splitTVShowCast.groupby('Actor').size()\
                                 .sort_values(ascending= False)\
                                 .reset_index(name = "Sum of TV Shows")
tvshowPerActor

```

```python
# Number of actors that have X number of movies
# Group by Sum of Movies
# count rows
# transform to DataFrame
# new column named "Actors"
# sort values in descending order by new column

moviePerActor.groupby("Sum of Movies")\
                         .size()\
                        .reset_index(name = "Actors")\
                        .sort_values("Actors", ascending = False)
                            
```

```python
tvshowPerActor.groupby("Sum of TV Shows")\
              .size()\
                .reset_index(name= "Actors")\
               .sort_values("Actors",ascending = False)


```

```python
#GENRE ANALYSIS
genres = noNanValuesDf.drop(columns = ["cast","release_year","rating","duration"])
genres
```

```python
movieGenres = genres[genres['type']=="Movie"]
movieGenres = movieGenres.drop(columns = ['type'])
movieGenres
```

```python
# splitMovieGenres = sPG
sPG = movieGenres.set_index('title')\
                .listed_in.str.split(",",expand = True)\
                .stack()\
                .reset_index('title')\
                .rename(columns = {0 :"Genres"})\
                .sort_values('Genres')[['Genres','title']]\
                .reset_index(drop=True)
sPG
```

```python
# Remove duplicate titles reset index
sPG = sPG.drop_duplicates(subset='title').set_index('Genres').reset_index()
sPG
```

```python
movieGenreCount = sPG.groupby('Genres').size()\
                     .sort_values(ascending = False)\
                     .reset_index(name = "Count")
movieGenreCount
```

```python
# TV Show Genres
tvshowGenres = genres[genres['type']=="TV Show"]
tvshowGenres = tvshowGenres.drop(columns = ['type'])
tvshowGenres
```

```python
#Splitting Show Genres
splittvshowGenres = tvshowGenres.set_index('title')\
                                .listed_in.str.split(",",expand = True)\
                                .stack()\
                                .reset_index('title')\
                                .rename(columns = {0 :"Genre"})[['Genre','title']]\
                                .sort_values(['Genre'])
                                
splittvshowGenres
```

```python
#REMOVE DUPLICATES
splittvshowGenres = splittvshowGenres.drop_duplicates(subset= 'title')

splittvshowGenres 
```

```python
tvshowCount = splittvshowGenres.groupby('Genre')\
                            .size()\
                            .sort_values(ascending = False)\
                             .reset_index(name = 'Count')
tvshowCount
```
